```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Compute the order counts per hour
hourly_order_counts = df_orders_products_merged.groupby('order_hour_of_day')['order_id'].count()
```

```python
# Compute the quartiles of the order counts
quartiles = np.quantile(hourly_order_counts, [0.25, 0.5, 0.75])
```

```python
# Define a function to label each hour based on its order count
def label_hour(row):
    if row['order_count'] <= quartiles[0]:
        return 'Fewest orders'
    elif row['order_count'] <= quartiles[1]:
        return 'Average orders'
    else:
        return 'Most orders'
```

```python
# Create a new column containing the labels
hourly_order_counts_df = hourly_order_counts.reset_index().rename(columns={'order_id': 'order_count'})
hourly_order_counts_df['busiest_period_of_day'] = hourly_order_counts_df.apply(label_hour, axis=1)
```

```python
# Merge the labels back into the original dataframe
df_orders_products_merged = pd.merge(df_orders_products_merged, hourly_order_counts_df[['order_hour_of_day', 'busiest_period_of_day']], on='order_hour_of_day', how='left')
```
